```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_clean = pd.read_csv("data/processed/monthly_housing_clean.csv")
df_clean.head(3)
```

## Visualize clean data

```python
plt.figure(figsize=(10, 4))

plt.plot(
    df_clean.index,
    df_clean["hpi"],
    linewidth=0.8
)

y_min = df_clean["hpi"].min()
y_max = df_clean["hpi"].max()
padding = (y_max - y_min) * 0.03

plt.ylim(y_min - padding, y_max + padding)

plt.title("Housing Price Index Over Time")
plt.xlabel("Date")
plt.ylabel("HPI")
plt.grid(True, alpha=0.3)

plt.show()
```

## Summary Statistic

```python
summary_stats = pd.DataFrame({
    "HPI": [
        df_clean["hpi"].mean(),
        df_clean["hpi"].std(),
        df_clean["hpi"].min(),
        df_clean["hpi"].quantile(0.25),
        df_clean["hpi"].median(),
        df_clean["hpi"].quantile(0.75),
        df_clean["hpi"].max(),
        df_clean["hpi"].std() / df_clean["hpi"].mean()
    ],
}, index=[
    "Mean",
    "Standard deviation",
    "Minimum",
    "First quartile",
    "Median",
    "Third quartile",
    "Maximum",
    "CV"
])

summary_stats_rounded = summary_stats.round(3)
summary_stats_rounded
```
